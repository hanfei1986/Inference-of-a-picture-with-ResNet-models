```python
from torchvision.models import resnet18, resnet34, resnet50, resnet101, resnet152
```

```python
net18 = resnet18(weights='ResNet18_Weights.DEFAULT')
net34 = resnet34(weights='ResNet34_Weights.DEFAULT')
net50 = resnet50(weights='ResNet50_Weights.DEFAULT')
net101 = resnet101(weights='ResNet101_Weights.DEFAULT')
net152 = resnet152(weights='ResNet152_Weights.DEFAULT')
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import torch
from torchvision import transforms
```

```python
img_raw = Image.open('picture.jpg').convert('RGB')
plt.imshow(img_raw)
plt.show()

# transforms.Compose([]) creates a pipeline of transformations to image
transform_eval = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),  # converts the resized image into a PyTorch tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
]) 
# mean and std values used here are precomputed values that 
# are commonly used for normalization in computer vision tasks. 通用于所有RGB图片？

img = transform_eval(img_raw)
```

```python
import json

with open("imagenet_idx_to_class.json", "r") as f:
    classes = json.load(f)
```

```python
net18.eval()  # not in train mode, don't update weights
pred = net18(img.unsqueeze(0)) # add a dimension for batch_size
len(pred[0])
```

```python
topK = 5
conf, ind = torch.topk(pred.squeeze(0), k=topK)
ind = ind.squeeze(0).numpy()
print('The input picture is classified to be')
for i in range(topK):
    print('\t[%s], with probability %.3f.'%
          (classes[ind[i]], torch.softmax(pred.squeeze(0), dim=0)[ind[i]].item()))
```
